```python
from sklearn.model_selection import train_test_split
df_train,df_test = train_test_split(df,test_size = 0.2,random_state=42,stratify=df['TenYearCHD'])
```

```python
import statsmodels.formula.api as smf
logit_reg = smf.logit("TenYearCHD~age", data = df_train)
logit_reg_results = logit_reg.fit()
print(logit_reg_results.summary())
```

```python
formula = "TenYearCHD ~ " + "+".join([col for col in df.columns if col!="TenYearCHD"])
print(formula)
```

```python
logit_all = smf.logit(formula, data=df_train)
logit_results = logit_all.fit()
print(logit_results.summary())
```

```python
logit_results.predict(df_test)
```

```python
df_test['PredictedCHD'] = logit_results.predict(df_test) > 0.5
```

```python
pd.crosstab(df_test['TenYearCHD'], df_test['PredictedCHD']) #model rarely predicts ppl get hearts disease
```

```python
'''
Receiver Operator characteristic (ROC) Curve

Helps us pick a threshold

True positive rate =  is the number of instances that were actually negative and correctly predicted as negative by the model.
False positive rate = is the number of instances that were actually negative but incorrectly predicted as positive by the model.

The True Positive Rate (TPR), also known as Sensitivity or Recall,
is a measure used to evaluate the performance of a binary classification model.
It measures the proportion of actual positive cases (true positives) that the model correctly identifies as positive.


The False Positive Rate (FPR) is a measure used to evaluate the performance of a binary classification model.
It measures the proportion of actual negative cases (true negatives) that the model incorrectly identifies as positive.


The AUC, or Area Under the Receiver Operating Characteristic Curve, is a commonly used metric to evaluate the performance of a binary classification model.
It quantifies the model's ability to distinguish between the two classes (positive and negative) across different threshold values for classifying instances.


'''
```

```python
from sklearn.metrics import roc_auc_score
df_test['PredictedProbTenYearCHD'] = logit_results.predict(df_test)
auc = roc_auc_score(df_test['TenYearCHD'], df_test['PredictedProbTenYearCHD'])
print(auc)
```
