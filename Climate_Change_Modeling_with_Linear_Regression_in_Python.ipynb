```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

training_set = pd.read_csv("ClimateChangeTrain-1.csv")
testing_set = pd.read_csv("ClimateChangeTest-1.csv")

trainTo2006 = training_set[training_set['Year'] <= 2006]
testFrom2007 = testing_set[testing_set['Year'] > 2006]

X = trainTo2006[['MEI', 'CO2', 'CH4', 'N2O', 'CFC11', 'CFC12', 'TSI', 'Aerosols']]
X = sm.add_constant(X)  # Adds a constant term to the predictor
y = trainTo2006['Temp']
model = sm.OLS(y, X).fit()
print(model.summary())

correlation_matrix = trainTo2006.corr()
print(correlation_matrix)

X_simplified = trainTo2006[['MEI', 'TSI', 'Aerosols', 'N2O']]
X_simplified = sm.add_constant(X_simplified)
newModel = sm.OLS(y, X_simplified).fit()
print(newModel.summary())


X_test = testFrom2007[['MEI', 'CO2', 'CH4', 'N2O', 'CFC11', 'CFC12', 'TSI', 'Aerosols']]
X_test = sm.add_constant(X_test)
pointsPredictions = model.predict(X_test)
SSE = np.sum((pointsPredictions - testFrom2007['Temp'])**2)
baseline = np.mean(trainTo2006['Temp'])
SST = np.sum((testFrom2007['Temp'] - baseline)**2)
R_squared = 1 - SSE/SST
print(f"Testing set R^2: {R_squared}")

# -----------------------------------------------------------------------------------------
# a).
# Question i)
#  linear regression equation =
#  Temp = (-124.594 + 0.0642 * MEI /n
#             + 0.00646 * CO2 /n
#             + 0.000124 * CH4 /n
#             - 0.0165 * N2O /n
#             - 0.00663 * CFC11 /n
#             + 0.00381 * CFC12 /n
#             + 0.0931 * TSI /n
#             - 1.5376 * Aerosols)

#-----------------------------------------------------------------------------------------
# Question ii)
# ii). R^2 = 0.7509
# MEI, CO2, CFC11, CFC12, TSI, and Aerosols are highly significant independent variables /n
# becuase those variables are all less than 0.05 the common significance threshold and standard p-value

#-----------------------------------------------------------------------------------------
# Question iii)
# All of the gas concentration variables in the data set /n
# are interconnected with human development - N2O and CFC-11 are n/
# positively correlated with other variables.

#-----------------------------------------------------------------------------------------
# Question iv)
# The independent variables that N2O is highly correlated with /n
#  (absolute correlation greater than 0.7) are: 'CO2', 'CH4', 'CFC12'

# The independent variables that CFC11 is highly correlated with /n
#  (absolute correlation greater than 0.7) are: 'CH4', 'CFC12'

# -----------------------------------------------------------------------------------------
# b).
# Question i)
# coefficient of N2O in this reduced model: 0.0253
# Previous model: -0.0165

#-----------------------------------------------------------------------------------------
# Question ii)
# Previous model R^2: 0.726

# -----------------------------------------------------------------------------------------
# c).
# R^2: 0.6274054133682756
# Testing set R^2: 0.6274054133682756. This tells us the value is loswer than both Model 1 and Model 2
# on the training data. This suggests that the model's performance has dropped
# when applied to new, unseen data (from 2007 onwards).
```
